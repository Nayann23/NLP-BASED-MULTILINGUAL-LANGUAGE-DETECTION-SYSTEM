```python
x = np.array(data["Text"])
y = np.array(data["language"])
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
```

```python
cv = CountVectorizer()
X = cv.fit_transform(x)


# print(Vectorizer.get_feature_names_out())  gives the unique words from the data

# print(Vectorized_Data.toarray()) 
# The output comes like this  First all unique words nikaalta hai
# ğŸ”¹ Unko sort karta hai (by default alphabetical)
# ğŸ”¹ Fir har sentence ke liye ek row banata hai
# ğŸ”¹ Har word ke liye column fix hota hai
# ğŸ”¹ Agar word present hai sentence mein â†’ us column mein 1, warna 0
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
print(X_train)
```

```python
y_test
```

```python
model = MultinomialNB()

model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
while True:

    UserInput = input("Enter Text To Detect Language: ")
    if UserInput == "Stop":
        break
    else:
        data = cv.transform([UserInput]).toarray()
        output = model.predict(data)
        print(f"Your text '{UserInput}.' Written in  {output} Language")
```
